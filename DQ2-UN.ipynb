```python
import numpy as np
```

```python
mu, sigma = 100, 15
iq_scores = np.random.normal(mu, sigma, 500)
```

```python
print(iq_scores)
```

```python
print(np.mean(iq_scores))
```

```python
print(np.std(iq_scores))
```

```python
import matplotlib.pyplot as plt
plt.hist(iq_scores)
plt.show()
```

```python
## This graph appears to be normally distributed.
```

```python
count, bins, ignored = plt.hist(iq_scores, 30, normed=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *np.exp( - (bins - mu)**2 / (2 * sigma**2) ),linewidth=2, color='r')
plt.show()
```
